```python
#Etudions les variables variables qualitatives : fréquences, répartitions, proportions.
import pandas as pd

# select qualitative vars
qual_vars = ['channel','device','company_size','sector','region','status']

results = {}
for col in qual_vars:
    eff = marketing_campaign[col].value_counts(dropna=False)
    pct = marketing_campaign[col].value_counts(normalize=True, dropna=False).mul(100).round(1)
    results[col] = pd.DataFrame({'Effectif': eff, 'Pourcentage': pct})

results
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Visualisations univariées
# =============================

# 1. Répartition des leads par canal
plt.figure()
marketing_campaign["channel"].value_counts().plot(kind="bar")
plt.title("Répartition des leads par canal")
plt.xlabel("Canal marketing")
plt.ylabel("Nombre de leads")
plt.show()

# 2. Répartition des leads par statut CRM
plt.figure()
marketing_campaign["status"].value_counts().plot(kind="bar")
plt.title("Répartition des leads par statut CRM")
plt.xlabel("Statut")
plt.ylabel("Nombre de leads")
plt.show()

# 3. Répartition des leads par device
plt.figure()
marketing_campaign["device"].value_counts().plot(kind="bar")
plt.title("Répartition des leads par device")
plt.xlabel("Device")
plt.ylabel("Nombre de leads")
plt.show()

# 4. Répartition des leads par secteur
plt.figure()
marketing_campaign["sector"].value_counts().plot(kind="bar")
plt.title("Répartition des leads par secteur")
plt.xlabel("Secteur")
plt.ylabel("Nombre de leads")
plt.show()

# 5. Répartition des leads par taille d’entreprise
plt.figure()
marketing_campaign["company_size"].value_counts().plot(kind="bar")
plt.title("Répartition des leads par taille d’entreprise")
plt.xlabel("Taille d’entreprise")
plt.ylabel("Nombre de leads")
plt.show()

# 6. Distribution des coûts de campagne
plt.figure()
marketing_campaign["cost"].plot(kind="hist", bins=5)
plt.title("Distribution des coûts de campagne")
plt.xlabel("Coût")
plt.ylabel("Fréquence")
plt.show()

# 7. Nombre de conversions par canal
plt.figure()
marketing_campaign.set_index("channel")["conversions"].plot(kind="bar")
plt.title("Nombre de conversions par canal")
plt.xlabel("Canal marketing")
plt.ylabel("Conversions")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ================================================================================

# =============================
# Bubble chart : Coût, conversions, impressions, clics, canal et statut
# Question métier : Quelle campagne maximise les conversions et l'engagement tout en optimisant le coût ?
# =============================
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=marketing_campaign, # Corrected: 'data' parameter is used to specify the DataFrame
    x="cost",                # Axe X : coût de la campagne
    y="conversions",         # Axe Y : conversions
    size="impressions",      # Taille des bulles : impressions
    hue="status",            # Couleur : statut du lead
    style="channel",         # Forme des points : canal
    sizes=(100,1000),
    alpha=0.7
)
plt.title("Performance des campagnes : Coût vs Conversions avec Engagement et Canal")
plt.xlabel("Coût de la campagne")
plt.ylabel("Nombre de conversions")
plt.legend(title="Statut / Canal", bbox_to_anchor=(1.05,1), loc=2)
plt.show()

# =============================
# Heatmap : Statut vs Canal vs Device vs Taille d’entreprise
# Question métier : Quels segments et canaux génèrent le plus de Clients ?
# ====================================================================

# Create a new column to indicate if a lead is a 'Client'
marketing_campaign['is_client'] = (marketing_campaign['status'] == 'Client').astype(int)

pivot_heatmap = pd.pivot_table(
   marketing_campaign,
    index="company_size",
    columns="channel",
    values="is_client",  # Use the new 'is_client' column as values
    aggfunc='sum'         # Simply sum the 'is_client' values
).fillna(0)

plt.figure(figsize=(8,6))
sns.heatmap(pivot_heatmap, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Nombre de Clients par taille d’entreprise et canal")
plt.xlabel("Canal marketing")
plt.ylabel("Taille d’entreprise")
plt.show()

# Impressions x Clics x Statut CRM
# Question métier : Comment la visibilité et l’engagement influencent le statut des leads ?
# ==================================================
plt.figure()
sns.scatterplot(
    data=marketing_campaign,
    x="impressions",
    y="clicks",
    hue="status",
    style="channel",
    s=100
)
plt.title("Clics vs Impressions selon le statut et le canal")
plt.xlabel("Nombre d'impressions")
plt.ylabel("Nombre de clics")
plt.legend(title="Statut / Canal", bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
# ==================================================
# Canal x Device x Statut CRM
# Question métier : Quels canaux et devices génèrent les leads les plus qualifiés ?
# ==================================================
pivot = pd.pivot_table(
    marketing_campaign,
    index="channel",
    columns="device",
    values="status",
    aggfunc=lambda x: (x=="Client").sum(),  # compter les clients
    fill_value=0
)

plt.figure()
pivot.plot(kind="bar")
plt.title("Nombre de Clients par canal et par device")
plt.xlabel("Canal marketing")
plt.ylabel("Nombre de Clients")
plt.legend(title="Device")
plt.show()
```

```python
# Inserer les 200 lignes de maniere aléatoire dans le dataframe marketing_campaign
import pandas as pd
import random
import numpy as np


# Listes de valeurs possibles (basées sur les données existantes)
channels = ["Emailing", "Facebook Ads", "LinkedIn", "Instagram Ads"]
devices = ["Mobile", "Desktop", "Tablet"]
company_sizes = ["1-10", "10-50", "50-100", "100-500"]
sectors = ["Tech", "Retail", "Finance", "Health", "Education"]
regions = ["IdF", "Hauts-de-France", "PAC", "ARA"]
statuses = ["MQL", "SQL", "Client"]
campaign_ids = ["CAMP01", "CAMP02", "CAMP03", "CAMP04"]

# Génération de 200 lignes aléatoires
rows = []
for i in range(200):
    row = {
        "lead_id": marketing_campaign["lead_id"].max() + i + 1,
        "date": f"2025-09-{random.randint(1,30):02d}",
        "channel": random.choice(channels),
        "device": random.choice(devices),
        "company_size": random.choice(company_sizes),
        "sector": random.choice(sectors),
        "region": random.choice(regions),
        "status": random.choice(statuses),
        "campaign_id": random.choice(campaign_ids),
        "cost": random.randint(800, 3500),
        "impressions": random.randint(20000, 100000),
        "clicks": random.randint(300, 3000),
        "conversions": random.randint(20, 300)
    }
    rows.append(row)

# Création d’un DataFrame avec les nouvelles lignes
df_random = pd.DataFrame(rows)

# Fusion avec le DataFrame original
df_extended = pd.concat([ marketing_campaign, df_random], ignore_index=True)

# 6. Vérification
print(df_extended.shape)
df_extended.head()

```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# --- 1) Charger le dataset ---
df = df_extended.copy()   # ou pd.read_csv("donnees_nettoyes_plus200.csv")

# --- 2) Calcul des KPI ---
df['CTR'] = df['clicks'] / df['impressions']
df['CVR'] = df['conversions'] / df['clicks']
df['CPC'] = df['cost'] / df['clicks']
df['CPA'] = df['cost'] / df['conversions']

# --- 3) Widgets interactifs ---
channel_dropdown = widgets.Dropdown(
    options=["Tous"] + sorted(df["channel"].unique()),
    value="Tous",
    description="Canal :",
)

device_dropdown = widgets.Dropdown(
    options=["Tous"] + sorted(df["device"].unique()),
    value="Tous",
    description="Device :",
)

status_dropdown = widgets.Dropdown(
    options=["Tous"] + sorted(df["status"].unique()),
    value="Tous",
    description="Statut :",
)

# --- 4) Fonction d’affichage KPI ---
def update_kpi(channel, device, status):

    dff = df.copy()

    if channel != "Tous":
        dff = dff[dff["channel"] == channel]
    if device != "Tous":
        dff = dff[dff["device"] == device]
    if status != "Tous":
        dff = dff[dff["status"] == status]

    kpis = dff[["CTR", "CVR", "CPC", "CPA"]].mean().round(4)

    display(kpis.to_frame("Valeur moyenne"))
    print(f"\nNombre d’enregistrements filtrés : {len(dff)}")

# --- 5) Interface interactive ---
ui = widgets.VBox([channel_dropdown, device_dropdown, status_dropdown])
out = widgets.interactive_output(update_kpi, {
    "channel": channel_dropdown,
    "device": device_dropdown,
    "status": status_dropdown
})

display(ui, out)
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = df_extended.copy()

df["CTR"] = df["clicks"] / df["impressions"]
df["CVR"] = df["conversions"] / df["clicks"]
df["CPC"] = df["cost"] / df["clicks"]
df["CPA"] = df["cost"] / df["conversions"]

# Graphique 1 : CTR et CVR par canal
fig1 = px.bar(
    df.groupby("channel")[["CTR","CVR"]].mean().reset_index(),
    x="channel", y=["CTR","CVR"],
    barmode="group", title="CTR & CVR par Canal"
)
fig1.show()

# Graphique 2 : CPC & CPA par canal
fig2 = px.bar(
    df.groupby("channel")[["CPC","CPA"]].mean().reset_index(),
    x="channel", y=["CPC","CPA"],
    barmode="group", title="CPC & CPA par Canal"
)
fig2.show()

# Graphique 3 : entonnoir impressions → clics → conversions
df_funnel = df.groupby("channel")[["impressions","clicks","conversions"]].sum().reset_index()
fig3 = px.funnel(df_funnel, x=["impressions","clicks","conversions"], y="channel",
                 title="Entonnoir Marketing par Canal")
fig3.show()
```

```python
# CTR / CVR / CPC / CPA par device
fig4 = px.bar(
    df.groupby("device")[["CTR","CVR","CPC","CPA"]].mean().reset_index(),
    x="device", y=["CTR","CVR","CPC","CPA"],
    barmode="group", title="KPI par Device"
)
fig4.show()

# Heatmap canal × device
pivot = df.pivot_table(index="channel", columns="device", values="conversions", aggfunc="sum")
fig5 = px.imshow(pivot, text_auto=True, color_continuous_scale="Blues",
                 title="Conversions par Canal et Device")
fig5.show()
```

```python
# Nombre par statut
status_counts = df["status"].value_counts().reset_index()
status_counts.columns = ['status_category', 'count_of_leads'] # Explicitly rename columns
fig6 = px.bar(
    status_counts,
    x="status_category",
    y="count_of_leads",
    title="Nombre de Leads par Statut",
    labels={"status_category":"Statut","count_of_leads":"Nombre"} # Update labels accordingly
)
fig6.show()

# KPI par statut
fig7 = px.bar(
    df.groupby("status")[["CTR","CVR","CPC","CPA"]].mean().reset_index(),
    x="status", y=["CTR","CVR","CPC","CPA"],
    barmode="group", title="KPI par Statut"
)
fig7.show()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="Dashboard Marketing SmartMarket", layout="wide")

# -------------------------------------------------------------
# Chargement des données
# -------------------------------------------------------------
df = pd.read_csv('df_extended.csv')# -------------------------------------------------------------
# Calcul des KPI (avec sécurisation)
# -------------------------------------------------------------
df["CTR"] = df["clicks"] / df["impressions"]
df["CVR"] = df["conversions"] / df["clicks"].replace(0, pd.NA)
df["CPC"] = df["cost"] / df["clicks"].replace(0, pd.NA)
df["CPA"] = df["cost"] / df["conversions"].replace(0, pd.NA)

# -------------------------------------------------------------
# Filtres
# -------------------------------------------------------------
st.sidebar.header("Filtres")

channel_f = st.sidebar.selectbox(
    "Canal", ["Tous"] + sorted(df["channel"].dropna().unique())
)
device_f = st.sidebar.selectbox(
    "Device", ["Tous"] + sorted(df["device"].dropna().unique())
)
status_f = st.sidebar.selectbox(
    "Statut", ["Tous"] + sorted(df["status"].dropna().unique())
)

df_f = df.copy()

if channel_f != "Tous":
    df_f = df_f[df_f["channel"] == channel_f]
if device_f != "Tous":
    df_f = df_f[df_f["device"] == device_f]
if status_f != "Tous":
    df_f = df_f[df_f["status"] == status_f]

st.subheader("Nombre d’enregistrements filtrés")
st.write(len(df_f))

# -------------------------------------------------------------
# KPI Cards
# -------------------------------------------------------------
st.subheader("KPI Clés")

col1, col2, col3, col4 = st.columns(4)

col1.metric("CTR Moyen", f"{df_f['CTR'].mean():.2%}")
col2.metric("CVR Moyen", f"{df_f['CVR'].mean():.2%}")
col3.metric("CPC Moyen (€)", f"{df_f['CPC'].mean():.2f}")
col4.metric("CPA Moyen (€)", f"{df_f['CPA'].mean():.2f}")

# -------------------------------------------------------------
# Performance par canal
# -------------------------------------------------------------
st.subheader("Performance par Canal")

kpi_by_channel = (
    df_f.groupby("channel")[["CTR", "CVR", "CPC", "CPA"]]
    .mean()
    .reset_index()
)

fig1 = px.bar(
    kpi_by_channel,
    x="channel",
    y=["CTR", "CVR"],
    barmode="group",
    title="CTR & CVR par Canal"
)
st.plotly_chart(fig1, use_container_width=True)

fig2 = px.bar(
    kpi_by_channel,
    x="channel",
    y=["CPC", "CPA"],
    barmode="group",
    title="CPC & CPA par Canal"
)
st.plotly_chart(fig2, use_container_width=True)

# -------------------------------------------------------------
# Funnel Marketing
# -------------------------------------------------------------
st.subheader(" Entonnoir Marketing")

df_funnel = (
    df_f.groupby("channel")[["impressions", "clicks", "conversions"]]
    .sum()
    .reset_index()
)

df_funnel_long = df_funnel.melt(
    id_vars="channel",
    value_vars=["impressions", "clicks", "conversions"],
    var_name="Étape",
    value_name="Volume"
)

fig3 = px.funnel(
    df_funnel_long,
    x="Volume",
    y="Étape",
    color="channel",
    title="Funnel Marketing par Canal"
)
st.plotly_chart(fig3, use_container_width=True)

# -------------------------------------------------------------
# Performance par Device
# -------------------------------------------------------------
st.subheader("Performance par Device")

kpi_by_device = (
    df_f.groupby("device")[["CTR", "CVR", "CPC", "CPA"]]
    .mean()
    .reset_index()
)

fig4 = px.bar(
    kpi_by_device,
    x="device",
    y=["CTR", "CVR", "CPC", "CPA"],
    barmode="group",
    title="KPI par Device"
)
st.plotly_chart(fig4, use_container_width=True)

# -------------------------------------------------------------
# Heatmap Canal x Device
# -------------------------------------------------------------
st.subheader("Heatmap Conversions : Canal × Device")

pivot = df_f.pivot_table(
    index="channel",
    columns="device",
    values="conversions",
    aggfunc="sum",
    fill_value=0
)

fig5 = px.imshow(
    pivot,
    text_auto=True,
    color_continuous_scale="Blues",
    title="Conversions par Canal et Device"
)
st.plotly_chart(fig5, use_container_width=True)
```
